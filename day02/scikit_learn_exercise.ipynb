### Heart Disease Dataset

The data we will work with is a collection of health measurements for 303 patietents admitted to the hospital. The class label indicates if the patient was ultimately diagnosed with heart disease.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9e/SinusRhythmLabels.svg/500px-SinusRhythmLabels.svg.png" width=500px>

Take a look at the file cleveland_heart_disease.csv

The columns represent 14 different attributes, including the class label. 

1. age: age of the patient in years       
2. sex: 1 = male; 0 = female
3. cp: chest pain type
 * Value 1: typical angina
 * Value 2: atypical angina
 * Value 3: non-anginal pain
 * Value 4: asymptomatic
4. trestbps: resting blood pressure (in mm Hg on admission to the 
        hospital)
5. chol: serum cholestoral in mg/dl
6. fbs:  fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false)    
7. restecg: resting electrocardiographic results
  * Value 0: normal
  * Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
  * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
8. thalach:   maximum heart rate achieved
9. exang:    exercise induced angina (1 = yes; 0 = no)
10. oldpeak: ST depression induced by exercise relative to rest
11. slope:   the slope of the peak exercise ST segment
 * Value 1: upsloping
 * Value 2: flat
 * Value 3: downsloping
12. ca: number of major vessels (0-3) colored by flourosopy
13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
14. class: 0=no heart disease; 1 = heart disease


You do not need to know what each attribute means, but it's useful to understand which of these are **categorical** features (i.e. they have multiple discrete category values) and which are **numerical**. 

The raw data was obtained from here: https://archive.ics.uci.edu/dataset/45/heart+disease

### Part 3: Visualizing the data in 3D (optional)
Next, we can try to visualize the data in 3D. Here is an example that illustrates how to create a 3D scatterplot using matplotlib using random data.

```python

from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(50) # list of all x coordinates
y = np.random.rand(50) # list of all y coordinates
z = np.random.rand(50) # List of all z coordinates

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create scatter plot
ax.scatter3D(x, y, z, color='red', marker='o')

# Labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('Basic 3D Scatter Plot')

plt.show()
```
